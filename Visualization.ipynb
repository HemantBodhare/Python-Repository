1.scatter plot--used for showing relationship between two or more variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#scatter plot
area=np.array([900,1200,3000,4000])  # independant--x axis
price=np.array([1.12,1.50,2.10,2.30]) # dependant--y axis
print(area)
print(price)
#plt.scatter(area,price)
#print(area)
#print(price)

plt.scatter(area,price,c="purple",marker="*",s=100)  # scatter(x axis var, y axis var)
plt.xlabel("Area in sq.ft",c="blue")
plt.ylabel("Price in cr",c="blue")
plt.title("Area vs Price",c="green")
plt.xticks(area)
plt.yticks(price)

plt.show()
```

```python
#line plot
plt.plot(area,price,marker=".")    # syntax--> plot(x axis , y axis)
#plt.scatter(area,price,c="purple",marker="*",s=100) 
plt.xlabel("Area in sq.ft",c="blue")
plt.ylabel("Price in cr",c="blue")
plt.title("Area vs Price",c="green")
plt.xticks(area)
plt.yticks(price)

plt.show()
```

```python
x=df_car['width']
y=df_car['price']
plt.scatter(x,y)
plt.xlabel("Width of car",c="blue")
plt.ylabel("Price of car",c="blue")
plt.title("Width vs Price",c="green")

plt.show()
```

Conclusion: Price of car is dense at bottom and scatter/distributed at the top, so price of car not only depend on width. It may depend on other columns/factors

```python
#histogram(frequency curve)
x=df_car['fuel-type']
plt.hist(x)
plt.xlabel('Fuel Type')
plt.ylabel('Number of cars')
plt.title('Fule type vs number of cars')

count=x.value_counts()
plt.yticks(count)
plt.show()
```

```python
#barplot(frequency curve)---vertical bar
count.plot(kind='bar') # syntax--> plot(kind='bar')  , plot(kind='barh')
plt.yticks(count)
plt.show()
```

```python
#barplot--horizontal bar
count.plot(kind='barh',fontsize=10)
plt.xticks(count)
plt.show()
```

```python
#pieplot
pie=df_car['body-style'].value_counts()
pie.plot(kind='pie',autopct='%.2f%%',explode=(0,0.05,0,0.1,0.1))  #autopct--value in percentage, explode--highlight the portaion anticlockwise

plt.show()
```

```python
#seaborn--high quality chart, 3D chart, directly work on dataframe
car_dict={
    'Area':area,
    'Price':price
}
df_car1=pd.DataFrame(car_dict)
df_car1
```

```python
x1=df_car1['Area']
y1=df_car1['Price']
sns.scatterplot(data=df_car1,x=x1,y=y1)  # calling seaborn function  syntax--->scatter(data=dataframe,x=x axis value, y=y axis value)
plt.show()     #calling matplotlib function
```

```python
#frequency curve using seaborn
#countplot function-->categorical data(string)
plt.figure(figsize=(7,5))  # avoid overlapping of numbers
sns.countplot(data=df_car,x='body-style',hue='body-style')   # for horizontal bar --> y='body-style'
count1=df_car['body-style'].value_counts()  #fixing y axis using exact number of cars
plt.yticks(count1)
plt.show()
```

```python
#combining two columns--side by side bar chart
sns.countplot(data=df_car,x='body-style',hue='fuel-type')
plt.show()
```

```python
#boxplot-->check outlier--data below lower range and above upper range
# syntax---seaborn.boxplot(x, y, hue, data)
X=df_car.iloc[:,:-1]#all the columns except price column
target=df_car.iloc[:,-1] #price
sns.boxplot(data=X,x=target,y='make')
plt.grid()
```

```python
#heatmap---works on matrix
corr_car=df_car1.corr()
sns.heatmap(corr_car,cmap='coolwarm') # coolwarm syntax--> heatmap(correlation variable, cmap=color)
plt.show()
```
