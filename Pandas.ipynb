```python
#creating new col
branch=['eclass','thane','kalyan','hydrabad','chennai']
df_stu_dict['Branch']=branch
df_stu_dict
```

```python
df_stu_dict['Percentage']=df_stu_dict['Marks']/200*100
df_stu_dict
```

```python
# apply condition on percentage col
# we can create function for grade allocation
def grade(per):          # percentage col  assigned to per variable
    if per>40:
        return 'A'
    elif per<=40:
        return 'B'

df_stu_dict['Grade']=df_stu_dict['Percentage'].apply(grade) # dataframe[col].apply(functionname)    # functioname(arguments)
df_stu_dict
        
```

```python
#value_counts--return total value avaiable in a col

df_stu_dict['Grade'].value_counts()
```
